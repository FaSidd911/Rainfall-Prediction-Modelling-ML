```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\PC\Python Programs\Rainfall\bombay.csv')
```

```python
pd.set_option("display.max_columns", None)
df.head()
```

```python
numerical_feature = [feature for feature in df.columns if df[feature].dtypes != 'O']
discrete_feature=[feature for feature in numerical_feature if len(df[feature].unique())<25]
continuous_feature = [feature for feature in numerical_feature if feature not in discrete_feature]
categorical_feature = [feature for feature in df.columns if feature not in numerical_feature]
print("Numerical Features Count {}".format(len(numerical_feature)))
print("Discrete feature Count {}".format(len(discrete_feature)))
print("Continuous feature Count {}".format(len(continuous_feature)))
print("Categorical feature Count {}".format(len(categorical_feature)))
```

```python
df = df.drop(columns = categorical_feature)
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot  as plt
import seaborn as sns
corrmat = df.corr()
plt.figure(figsize=(20,20))
g=sns.heatmap(corrmat,annot=True)
```

```python
for feature in continuous_feature:
    data=df.copy()
    sns.distplot(df[feature])
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.figure(figsize=(15,15))
    plt.show()
```

```python
for feature in continuous_feature:
    data=df.copy()
    sns.boxplot(data[feature])
    plt.title(feature)
    plt.figure(figsize=(15,15))
```

```python
df.loc[df['precipMM']>=1,'precipMM']= 1
df.loc[(df['precipMM']>=0.1) & (df['precipMM']<0.5) ,'precipMM']=0.1
df.loc[(df['precipMM']>=0.5) & (df['precipMM']<1) ,'precipMM']=0.5
```

```python
continuous_feature.remove('precipMM')
```

```python
for feature in continuous_feature:
    IQR = df[feature].quantile(0.75) -  df[feature].quantile(0.25) 
    lower_bridge=df[feature].quantile(0.25)-(IQR*1.5)
    upper_bridge=df[feature].quantile(0.75)+(IQR*1.5)
    df.loc[df[feature]>=upper_bridge,feature]=upper_bridge
    df.loc[df[feature]<=lower_bridge ,feature]=lower_bridge
```

```python
for feature in discrete_feature:
    data=df.copy()
    sns.boxplot(data[feature])
    plt.title(feature)
    plt.figure(figsize=(15,15))
```

```python
import scipy.stats as stats
for feature in continuous_feature:
    print(feature)
    plt.figure(figsize=(15,6))
    plt.subplot(1, 2, 1)
    df[feature].hist()
    plt.subplot(1, 2, 2)
    stats.probplot(df[feature], dist="norm", plot=plt)
    plt.show()
```

```python
for feature in discrete_feature:
    print(feature)
    plt.figure(figsize=(15,6))
    plt.subplot(1, 2, 1)
    df[feature].hist()
    plt.subplot(1, 2, 2)
    stats.probplot(df[feature], dist="norm", plot=plt)
    plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['precipMM_enc']= label_encoder.fit_transform(df['precipMM'])
```

```python
df.to_csv("preprocessed_1.csv", index=False)
```

```python
from sklearn import preprocessing
import scipy.stats as stats
from sklearn.model_selection import train_test_split
from collections import Counter
from imblearn.over_sampling import SMOTE
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
import joblib
```

```python
X = df.drop(["precipMM","precipMM_enc"], axis=1)
Y = df["precipMM_enc"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size =0.2, random_state = 0)
```

```python
from imblearn.over_sampling import RandomOverSampler
from collections import Counter
sm=RandomOverSampler(random_state=0)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train)
print("The number of classes before fit {}".format(Counter(y_train)))
print("The number of classes after fit {}".format(Counter(y_train_res)))
```

```python
rf=RandomForestClassifier()
rf.fit(X_train_res,y_train_res)
```

```python
y_pred1 = rf.predict(X_test)
```

```python
print(confusion_matrix(y_test,y_pred1))
```

```python
print(accuracy_score(y_test,y_pred1))
```

```python
print(classification_report(y_test,y_pred1))
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier()
xgb.fit(X_train_res, y_train_res)
```

```python
y_pred2 = xgb.predict(X_test)
print(confusion_matrix(y_test,y_pred2))
print(accuracy_score(y_test,y_pred2))
print(classification_report(y_test,y_pred2))
```
